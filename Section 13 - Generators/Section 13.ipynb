# Python Generators

Range() itself is a generator

```python
def create_cubes(n):
    
    for x in range(n):
        yield x**3
    
```

```python
list(create_cubes(10))
```

```python
for x in create_cubes(10):
    print(x)
```

```python
def gen_fibon(n):
    a = 1
    b = 1
    for i in range(n):
        yield a
        a,b = b,a+b
```

```python
gen_fibon(10)
```

```python
for num in gen_fibon(10):
    print(num)
```

```python
def simple_gen():
    for i in range(3):
        yield i
```

```python
for num in simple_gen():
    print(num)
```

```python
g = simple_gen()
```
